```python
#Code Implementation Task: Implement code to fill the missing data (impute) in daily_vaccinations column per country with the minimum daily vaccination number of relevant countries.  
#Note: If a country does not have any valid vaccination number yet, fill it with “0” (zero). 

df = pd.read_csv('https://www.piworks.net/Upload/Document/Original/country_vaccination_stats.csv')

df['daily_vaccinations'] = df['daily_vaccinations'].fillna(0)


min_daily_vaccinations_by_country = df.groupby('country')['daily_vaccinations'].min()


df['daily_vaccinations'] = df.apply(lambda x: min_daily_vaccinations_by_country[x['country']] if x['daily_vaccinations'] == 0 else x['daily_vaccinations'], axis=1)

df
```
